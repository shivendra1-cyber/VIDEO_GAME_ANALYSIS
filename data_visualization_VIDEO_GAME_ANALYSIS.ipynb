# How to choose the right plot?

# 1. How many variables are involved in answering that question?
# 2. Whether they are continous or categorical variable?

# Univariate Data Visualisation
  # N --> Histogram, cdf,boxplot,pdf,histogram,violinplot,density
  # C --> XXXX, XXXX, XXX
# Bivariate Data Visualisation (Multi-variate)
  # N, N --> Scatterplot
  # N, C -->
  # C, C -->
# Multi-variate Data Visualisation
  # N, N, N
  # C, C, C
  # C, C, N
  # N, N, C

```python
cat_counts = data["Genre"].value_counts()
cat_counts
```

```python
x_val = cat_counts.index
y_val = cat_counts
plt.bar(x_val, y_val)
```

```python
plt.figure(figsize=(12,8))
plt.bar(x_val, y_val)
```

```python
plt.figure(figsize=(10,5))
plt.bar(x_val, y_val, width=0.5, color="red")
plt.xticks(rotation=90, fontsize=10)
plt.xlabel("Genre", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.title("Number of games per category")
```

```python
data["Genre"].value_counts().index
```

```python
sns.countplot(data=data, x="Genre", 
              order=data["Genre"].value_counts().index,
              color="orange")
plt.xticks(rotation=90)
plt.title("Number of games per category")
plt.show()
```

```python
sns.countplot(data=data, x="Genre",order=data["Genre"].value_counts().index,
              color="orange")
plt.xticks(rotation=90)
```

# Bivariate - Num, Num

```python
# Ice Hockey --> North American Sales of Ice Hockey for every year
# Year
# NA
ih = data.loc[data['Name']=='Ice Hockey']
bb = data.loc[data['Name']=='Baseball']
# for d in [ih, bb]:
#   sns.lineplot(data=d, .......)
sns.lineplot(data=ih, x="Year", y="NA_Sales", color="pink")
sns.lineplot(data=bb, x="Year", y="NA_Sales", color="blue")
plt.legend(["Ice Hockey", "Baseball"], loc="lower left")
plt.grid()
plt.show()
```
